```python
population_2021 = df["2021"].dropna()
df
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.hist(population_2021, bins=20)
plt.title("Population Distribution Across Countries (2021)")
plt.xlabel("Population")
plt.ylabel("Number of Countries")
plt.show()
```

```python
top10 = df[["Country Name", "2021"]].dropna().sort_values(
    by="2021", ascending=False
).head(10)

plt.figure(figsize=(10,5))
plt.bar(top10["Country Name"], top10["2021"])
plt.title("Top 10 Most Populous Countries (2021)")
plt.xlabel("Country")
plt.ylabel("Population")
plt.xticks(rotation=45)
plt.show()
```
